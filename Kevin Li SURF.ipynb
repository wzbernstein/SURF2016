```python
import numpy 
import matplotlib.pylab as plt
import scipy.cluster.hierarchy as sch
import pylab
```

```python
%matplotlib inline
def prepMatrix (data, label, titleText):
    matrix = numpy.matrix(data)
    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(1,1,1)
    #width, height = matrix.shape
    ax.set_aspect('equal')
    plt.imshow(matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.xticks(range(26), label, rotation='vertical')
    plt.yticks(range(26), label)
    plt.title(titleText)
    plt.colorbar()
    plt.show()
    
def clusterMatrix (data, label):
    data1 = numpy.matrix(data)
    # Fist, set up the figure
    fig = pylab.figure(figsize=(10,10))
    # fig.patch.set_visible(False) # plot on transparent background
    ax1 = fig.add_axes([0.01,0.01,0.5,0.5], # left, bottom, width, height
                       {'width':0.05, 
                        'edgecolor':'k',
                        'backgroundcolor':'w'}) 
    # values of left...height are in fractions of figure width & height
    spines_to_remove = ['left', 'bottom', 'right', 'top']
    for spine in spines_to_remove:
        ax1.spines[spine].set_visible(False)
        ax1.set_xticks([])
        ax1.set_yticks([])
    Y = sch.linkage(data1, method='complete')
    Z1 = sch.dendrogram(Y, orientation='right')
    Z2 = sch.dendrogram(Y, orientation='right', labels = label, leaf_font_size=9)
    axmatrix = fig.add_axes([0.51,0.01,0.5,0.5])
    idx1 = Z1['leaves']
    idx2 = Z2['leaves']
    data1 = data1[idx1,:]
    data1 = data1[:,idx2]
    im = axmatrix.matshow(data1, aspect='auto', origin='lower', cmap=pylab.cm.Blues)
    axmatrix.set_xticks([])
    axmatrix.set_yticks([])
    reorderedWords = Z2['ivl']
    axmatrix.set_xticklabels(reorderedWords[:-1])
    plt.show(fig)
```
